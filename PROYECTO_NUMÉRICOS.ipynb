```python
#Programa que implementa el metodo de Runge-Kutta de orden 4 para sistemas de 2 ecuaciones

#Ingresar los limites superior e inferior del intervalo [a,b]
a = float(input("Ingrese el limite inferior (a): "))
b = float(input("Ingrese el limite superior (b): "))

#Ingresar el tamaño del intervalo
N = int(input("Ingrese el entero N: "))

#Ingresar las condiciones iniciales y1,2(a) = w1,2
w1 = float(input("Ingrese el valor de la condicion inicial y1(a) = "))
w2 = float(input("Ingrese el valor de la condicion inicial y2(a) = "))

#Definimos la solucion exacta de la ecuacion diferencial 1
import math
def g1(t):
	return 0.2*math.exp(2*t)*(math.sin(t)-2*math.cos(t))

#Definimos la solucion exacta de la ecuacion diferencial 2
def g2(t):
	return 0.2*math.exp(2*t)*(4*math.sin(t)-3*math.cos(t))

#Arreglo inicial del metodo
h = (b-a)/N
t = a

#Arreglo inicial de la tabla
print("---------------------------------------------------------------------")
print(" t", "  y(t)=u1(t)" , "     w1 ","    y'(t)=u2(t)", "    w2",\
 "    |y(t)-w1|", "  |y'(t)-w2|")
print("---------------------------------------------------------------------")
print(format(t,'.1f'), format(g1(t),'.8f'),format(w1,'.8f'), format(g2(t),'.8f'),\
	format(w2,'.8f'), format(abs(g1(t)-w1),'.8f'), format(abs(g2(t)-w2),'.8f'))

#Definimos la funcion f1(t,w) = y1'
def f1(t,u1,u2):
	return u2

#Definimos la funcion f2(t,w) = y2'
def f2(t,u1,u2):
	return math.exp(2*t)*math.sin(t)-2*u1+2*u2

#Inicio del metodo
for i in range (1,N+1):
	k11 = h*f1(t,w1,w2)
	k12 = h*f2(t,w1,w2)

	k21 = h*f1(t+h/2, w1+k11/2, w2+k12/2)
	k22 = h*f2(t+h/2, w1+k11/2, w2+k12/2)

	k31 = h*f1(t+h/2, w1+k21/2, w2+k22/2)
	k32 = h*f2(t+h/2, w1+k21/2, w2+k22/2)

	k41 = h*f1(t+h, w1+k31, w2+k32)
	k42 = h*f2(t+h, w1+k31, w2+k32)

	w1 = w1+(k11+2*k21+2*k31+k41)/6
	w2 = w2+(k12+2*k22+2*k32+k42)/6

	t = round(a+i*h,2)
#Resultados para la tabla
	print(format(t,'.1f'), format(g1(t),'.8f'),format(w1,'.8f'), format(g2(t),'.8f'),\
		format(w2,'.8f'), format(abs(g1(t)-w1),'.8f'), format(abs(g2(t)-w2),'.8f'))
print("---------------------------------------------------------------------")
```

```python
import numpy as np
import math

# Definir el modelo general
def Model(x,params):
    M = params["masa_M"]
    m = params["masa_m"]
    alpha = params["alpha"]
    g = params["g"]

    xdot = np.array(\
    [x[2],\
    x[3],\
    g*math.sin(alpha)/(2-(m*math.cos(alpha)**2)/(m+M)),\
    -(m*g*math.sin(alpha)*math.cos(alpha))/(2*(m+M)-m*math.cos(alpha)**2)])

    return xdot

#RK4 Method
def RungeKutta4(f,x0,t0,tf,dt):

    t = np.arange(t0,tf,dt)
    nt = t.size

    nx = x0.size
    x = np.zeros((nx,nt))

    x[:,0] = x0

    for k in range(nt - 1):
        k1 = dt*f(t[k],x[:,k])
        k2 = dt*f(t[k] + dt/2, x[:,k] + k1/2)
        k3 = dt*f(t[k] + dt/2, x[:,k] + k2/2)
        k4 = dt*f(t[k] + dt, x[:,k] + k3)

        dx = (k1 + 2*k2 + 2*k3 + k4)/6

        x[:,k+1] = x[:,k] + dx

    return x, t

#Define problem

params = {"masa_M":10.0, "masa_m":8/3, "alpha":0.4886, "g":9.8}

f = lambda t, x : Model(x,params)

# condiciones iniciales b, a, bdot, adot
x0 = np.array([5,0,-5,0])

#Solve the Diff Eq

t0 = 0
tf = 40
dt = 0.01

x,t = RungeKutta4(f,x0,t0,tf,dt)
archivo1 = open('sol_pos_y_vel.txt', 'w')
archivo2 = open('sol_energía.txt', 'w')
l = 20.0
r = 0.1
M = params["masa_M"]
m = params["masa_m"]
alpha = params["alpha"]
g = params["g"]
for i in range(t.size-1):
    ener_cinetica = m*x[2,i]**2 + (m+M)*x[3,i]**2/2 + m*x[3,i]*x[2,i]*math.cos(alpha)
    ener_potencial = m*g*(r*math.cos(alpha) + (l-x[2,i])*math.sin(alpha))
    const = (M+m)*x[3,i]+m*x[2,i]*math.cos(alpha)
    print(t0+i*dt,x[0,i],x[1,i],x[2,i],x[3,i],file = archivo1)
    print(t0+i*dt,ener_cinetica,ener_potencial,ener_cinetica+ener_potencial,const,file = archivo2)

archivo1.close()
archivo2.close()

```
